# Barriers and Tunneling

## Barriers
### $E > V_0$

We will start with the definition of a potential barrier, considering a particle moving in the positive x direction toward it.
The total energy of the particle, E, exceeds the potential energy of the barrier it approaches. $E = K + V$, where $E > V$. As the particle moves closer to the barrier, eventually moving past it, classical mechanics says that the particle's kinetic energy goes down, as $K = mv^2 / 2 = E - V_0$. When the particle passes the barrier, the particle's velocity goes back to how it was originally. However, quantum mechanics informs us that the particle exhibits wave-like characteristics. Using our knowledge in optics, we can infer that when the particle is in the potential barrier, there is a change in the particle's wavenumber. We can use the analogy of light moving from air, through a medium like glass, and its wavelength changes due to the glass's index of refraction.

```python
# Parameters
V0 = 4
L = 3
E = 6

# x values
x = np.linspace(-4, 7, 1000)

V = np.piecewise(x,[x < 0, (x >= 0) & (x <= L), x > L],[0, V0, 0])

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, V, color='deepskyblue', linewidth=3)

ax.hlines(E, xmin=-4, xmax=7,colors='black',linestyles='dashed')
ax.annotate('',xy=(-2.5, E),xytext=(-3.5, E),arrowprops=dict(arrowstyle='->',color='deepskyblue',lw=2))

# Labels
ax.text(-4, E - 1, 'Particle', fontsize=18)
ax.text(-0.8, E + 0.2, r'$E_0$', fontsize=18)
ax.text(0.2, V0 + 0.2, r'$V_0$', fontsize=18)
ax.text(L - 0.1, -1, r'$L$', fontsize=18)
ax.text(L-0.05, -0.1, r'$|$', fontsize = 12)
ax.text(-0.4, -0.7, r'$0$', fontsize = 18)
ax.text(-2.5, -1.3, 'Region I', fontsize=18)
ax.text(0.3, -1.3, 'Region II', fontsize=18)
ax.text(4.2, -1.3, 'Region III', fontsize=18)
ax.text(7.2, -0.3, r'$x$', fontsize=18)
ax.text(-1.1, 7.2, r'$V(x)$', fontsize=18)
ax.axhline(0, color='black', linewidth=1.5)
ax.axvline(0, color='black', linewidth=1.5)
ax.set_xlim(-4, 7)
ax.set_ylim(-2, 8)
ax.set_xticks([])
ax.set_yticks([])

# Remove box
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
# Parameters
x_air = np.linspace(-6, 0, 1000)
x_glass = np.linspace(0, 6, 1000)

# Wavelengths
lambda_air = 2.0
lambda_glass = 1.0 

y_air = np.sin(2 * np.pi * x_air / lambda_air)
y_glass = np.sin(2 * np.pi * x_glass / lambda_glass)

fig, ax = plt.subplots(figsize=(10, 4))

# Plot waves
ax.plot(x_air, y_air, linewidth=2, label='Air')
ax.plot(x_glass, y_glass, linewidth=2, label='Glass')

# Boundary line
ax.axvline(0, linestyle='--', linewidth=1.5)
ax.text(-4.5, 1.3, 'Air', fontsize=16)
ax.text(2.5, 1.3, 'Glass', fontsize=16)
ax.annotate('',xy=(5.5, 0),xytext=(-5.5, 0),arrowprops=dict(arrowstyle='->', lw=2))
ax.annotate('',xy=(-4, -1.2),xytext=(-2, -1.2),arrowprops=dict(arrowstyle='<->', lw=1.5))
ax.text(-3.4, -1.5, r'$\lambda_{\mathrm{air}}$', fontsize=14)
ax.annotate('',xy=(1, -1.2),xytext=(2, -1.2),arrowprops=dict(arrowstyle='<->', lw=1.5))
ax.text(1.1, -1.5, r'$\lambda_{\mathrm{glass}}$', fontsize=14)
ax.set_xlim(-6, 6)
ax.set_ylim(-2, 2)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)
plt.title("Light Wave Entering Glass and Changing Wavelength", fontsize=16)
plt.show()
```

```python
V0 = 3
L = 2
x_barrier = [-7, 0, 0, L, L, 7]
y_barrier = [0, 0, V0, V0, 0, 0]

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_barrier, y_barrier, color='deepskyblue', linewidth=3)
ax.axhline(0, color='black', linewidth=1.5)
ax.axvline(0, color='black', linewidth=1.5)

# --- Incident Wave ---
x_inc = np.linspace(-5, -1.5, 400)
y_inc = 0.18 * np.sin(10 * x_inc) + 5

ax.plot(x_inc, y_inc, color='deepskyblue', linewidth=2)

# Arrow for incident wave
ax.annotate(
    '',
    xy=(-0.75, 5),
    xytext=(-1.5, 5),
    arrowprops=dict(arrowstyle='->',
                    color='deepskyblue',
                    lw=2))
# --- Reflected Wave ---
x_ref = np.linspace(-5, -1.5, 400)
y_ref = 0.18 * np.sin(10 * x_ref) + 3.5

ax.plot(x_ref, y_ref, color='deepskyblue', linewidth=2)

# Arrow for reflected wave
ax.annotate(
    '',
    xy=(-5.75, 3.5),
    xytext=(-5, 3.5),
    arrowprops=dict(arrowstyle='->',
                    color='deepskyblue',
                    lw=2))
# --- Transmitted Wave ---
x_trans = np.linspace(3, 6, 400)
y_trans = 0.18 * np.sin(10 * x_trans) + 5

ax.plot(x_trans, y_trans, color='deepskyblue', linewidth=2)

# Arrow for transmitted wave
ax.annotate(
    '',
    xy=(6.75, 5),
    xytext=(6, 5),
    arrowprops=dict(arrowstyle='->',
                    color='deepskyblue',
                    lw=2))
# Labels
ax.text(-5.3, 5.5, 'Incident', fontsize=20)
ax.text(-5.4, 2.8, 'Reflected', fontsize=20)
ax.text(3.1, 3.6, 'Transmitted', fontsize=20)

ax.text(-1.3, 6.3, r'$V(x)$', fontsize=20)
ax.text(6.2, -00.5, r'$x$', fontsize=20)
ax.text(0.5, 5, r'$E_0$', fontsize=20)
ax.text(-0.25, 5, r'$-$', fontsize=20)
ax.text(-0.6, -0.5, r'$0$', fontsize=18)
ax.text(L - 0.05, -0.8, r'$L$', fontsize=18)

# Styling
ax.set_xlim(-6, 7)
ax.set_ylim(-1, 7)

ax.set_xticks([])
ax.set_yticks([])

# Remove frame
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

Because we consider the wave behavior of particles with potential barriers, we can assume similar behavior. This means that 
the wave function will have a wave going in, a wave coming out, and a wave bouncing back from the potential barrier. Our
knowledge in quantum mechanics tells us that all of the light will be transmitted when $E >> V_0$, and all will be reflected when $E << V_0$.
This is where solving Scrodinger's Equations for the potentials of the three regions comes in handy for finding the wave functions.$$Region\space 1: \hspace{3cm} (X < 0) \hspace{3cm} V = 0 \hspace{3 cm} \frac{d^2\psi_I}{dx^2} + \frac{2m}{\hbar^2} E  \psi_I = 0$$
$$Region\space 2: \hspace{3cm} (0 < X < L) \hspace{3cm} V = V_0 \hspace{3 cm} \frac{d^2\psi_II}{dx^2} + \frac{2m}{\hbar^2} (E - V_0) \psi_II = 0$$
$$Region\space 3: \hspace{3cm} (X > L) \hspace{3cm} V = 0 \hspace{3 cm} \frac{d^2\psi_III}{dx^2} + \frac{2m}{\hbar^2} (E) \psi_III = 0$$The wavefunctions are as follows:
$$ \psi_I = A e^{ik_Ix} + B e^{-ik_Ix}$$
$$ \psi_{II} = C e^{ik_{II}x} + D e^{-ik_{II}x}$$
$$ \psi_{III} = F e^{ik_{III}x} + G e^{-ik_{III}x}$$All of the positive exponents represent transmitted waves, while the negative exponents represent reflected waves. In region 1, there
are waves moving in both  $+\hat{x}$ and $-\hat{x}$ directions, while in region 3, there are only transmitted waves, moving in the $+\hat{x}$ direction. Thus, the summarization of the wave functions: $$\psi_{I} \space (incident)=\space Ae^{ik_Ix}$$ $$\psi_I\space(reflected)
=\space Be^{-ik_Ix}$$ $$\psi_{III}\space(transmitted)=\space Fe^{ik_{III}x}$$ We can now relate the probabilities of the wave being transmitted or reflected by the potential barrier to the complex conjugates of the wave functions expressed above:

$$ R = \frac{\lvert {\psi_I\space (reflected)\rvert}^2} {\lvert {\psi_I\space (incident)\rvert}^2 } = \frac{B^{\star} B}{A^{\star} B}$$
$$ T = \frac{\lvert {\psi_{III}\space (transmitted)\rvert}^2} {\lvert {\psi_I\space (incident)\rvert}^2 } = \frac{F^{\star} F}{A^{\star} A}$$These are the probabilities of the particle being reflected (R) or transmitted (T). Because of the math behind probabilities,
we know that R + T = 1. Applying the boundary conditions of $ x \rightarrow \pm\infty$, $x = 0$, and $x = L$ gives us a relationship between all the coefficients of the different wave functions. The resulting Transmition probability is: $$T = [1 + \frac{V_0^2\sin^2(k_{II}L)}{4E(E-V_0)}]^{-1}$$  However, there is a situation when T = 1 itself, when $k_{II}L = n\pi$, $n$ being an integer. There is also a chance that the particle gets reflected at both $ x = 0$ and $x = L$. The path difference between the two beams, now moving in the $-\hat{x}$ direction, is $2L$. If the sum of all the wavelengths in the barrier is $ = 2L$, the incident and reflected wave functions are completely out of phase with each other and cancel each other out. 

## $E < V_0$: Tunneling

Now we must consider a situation in which the particle's energy is less than the potential barrier's height. This is where we see the phenomenon of **tunneling**. The following diagram shows what classical mechanics says in this situation, in which the particle is **always** reflected. The code is relatively the same, just lowering the particle and adding a reflected beam of light.

```python
# Parameters
V0 = 4
L = 3
E = 3
R = E - 2
# x values
x = np.linspace(-4, 7, 1000)

V = np.piecewise(x,[x < 0, (x >= 0) & (x <= L), x > L],[0, V0, 0])

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, V, color='deepskyblue', linewidth=3)

ax.hlines(E, xmin=-4, xmax=0,colors='black',linestyles='dashed')
ax.annotate('',xy=(-2.5, E),xytext=(-3.5, E),arrowprops=dict(arrowstyle='->',color='deepskyblue',lw=2))
ax.hlines(R, xmin=-4, xmax=0, colors='black', linestyles='dashed')
ax.annotate('',xy=(-2.5, R), xytext=(-3.5, R), arrowprops=dict(arrowstyle='<-', color='red', lw=2))

# Labels
ax.text(-4, E + 1, 'Particle', fontsize=18)
ax.text(-4, R + 1, 'Reflected', fontsize = 18)
ax.text(-0.8, E + 0.2, r'$E_0$', fontsize=18)
ax.text(0.2, V0 + 0.2, r'$V_0$', fontsize=18)
ax.text(L - 0.1, -1, r'$L$', fontsize=18)
ax.text(L-0.05, -0.1, r'$|$', fontsize = 12)
ax.text(-0.4, -0.7, r'$0$', fontsize = 18)
ax.text(-2.5, -1.3, 'Region I', fontsize=18)
ax.text(0.3, -1.3, 'Region II', fontsize=18)
ax.text(4.2, -1.3, 'Region III', fontsize=18)
ax.text(7.2, -0.3, r'$x$', fontsize=18)
ax.text(-1.1, 7.2, r'$V(x)$', fontsize=18)
ax.axhline(0, color='black', linewidth=1.5)
ax.axvline(0, color='black', linewidth=1.5)
ax.set_xlim(-4, 7)
ax.set_ylim(-2, 8)
ax.set_xticks([])
ax.set_yticks([])

# Remove box
for spine in ax.spines.values():
    spine.set_visible(False)
plt.title("Classical Behavior of E < V" )
plt.show()
```

However, a remarkable event occurs when we apply our knowledge in quantum mechanics. There is a probability that the particle would actually
move through the potential barrier, and come out the other side. This is the phenomenon of **Tunneling**. 

Analyzing the wave functions for the respective regions are the same except for one tweak to Region II, where $\psi_{II} = Ce^{\kappa x} + De^{-\kappa x}$, where $$\kappa = \frac{\sqrt{2m(V_0 - E)}} {\hbar}$$ The parameter kappa is a positive integer as $E < V_0$, thus $E - V_0$ is a positive difference. by changing the parameter of the transmission and reflection from $ik_{II}$ to $\kappa$, we see the probability of the particle continuing in the $+\hat{x}$ direction. The transmission probability changes as well: $$T = [1+\frac{(V_0)^2 \sinh^2 (\kappa L)}{4E(V_0-E)}]^{-1}$$Notice how the transmission probability changed from a sin function to a hyperbolic sin function. when $\kappa L$ is $>>1$, the probability turns into: $$T = 16\frac{E}{V_0}(1 - \frac{E}{V_0}) e^{-2\kappa L}$$ We can use the Heisenberg uncertainty principle to understand tunneling more thoroughly. The wave function $\psi_{II}$ has a dominating term of $e^{-\kappa x}$, and $(\lvert\psi_{II}\rvert)^2 = e^-2\kappa x$. Over the interval of $\Delta x = \kappa^-1$, the probability decreases by quite a lot ($e^{-2} = 0.14$. We now use the Heisenberg uncertainty principle, $\Delta p \Delta x \geq \hbar$, thus, $\Delta p \geq \frac{\hbar}{\Delta x} = \hbar \kappa$. By this, we know that the minimum kinetic energy is: $$K_{min} = \frac{\Delta p ^2}{2m} = \frac{\pi^2 \kappa^2}{2m} = V_0 - E$$ Meaning that by the Heisenberg uncertainty principle, and quantum mechanics, we have calculated the minimum kinetic energy required for the wavefunction to move through the potential barrier and go through the other side to be equal to the negative of the Heisenberg uncertainty princple!

### Potential Wells

For a particle of $E > 0$, and $V = -V_0$ between $x = 0$ and $x = L$, we see a "negative potential barrier," or a **potential well**. 
Under classical mechanics, the kinetic energy of the particle (i.e., the particle's speed) would increase as $K = 
\frac{mv^2}{2} = E + V_0$ If the width of the well is equal to the half-integral or integral sum of the wavelengths, then the reflected
wave is out-of-phase with the incident wave, causing potential cancelations. These cancellations can lead to near-complete transmission or reflection.

## Alpha-Particle Decay

There are many elements heavier than lead that emit alpha particles naturally, where the emission rates vary heavily, but the energies vary by only $4 to 8 MeV$. Inside the nucleus, the alpha-particles feel a short-range nuclear force, and outside of the nucleus, a powerful, longer-ranged Coulomb force. The coulomb potential energy barrier can be many times the magnitude of the energy of the alpha particle, making the particle 'trapped' inside the nucleus. However, the particle can tunnel through the barrier, which leads to varying amounts of emission rates. The probability, being the exponential $e^{-2\kappa L}$, means that any small change in the barrier creates a large fluctuation in the amount of alpha-particles.

```python
fig, ax = plt.subplots(figsize=(8, 5))
r_N = 1.0
L = 4.0
r = np.linspace(r_N, r_N + L + 1, 400)
V = 4 / r
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Nuclear well (vertical drop and flat bottom)
ax.plot([0, r_N], [-5, -5], color='deepskyblue', linewidth=3)
ax.plot([r_N, r_N], [-5, 4], color='deepskyblue', linewidth=3)

ax.plot(r, V, color='deepskyblue', linewidth=3)

E_alpha = 0.8
ax.hlines(E_alpha, 0, r_N + L, colors='gray', linewidth=1.5)

V_C = 4 / r_N
ax.hlines(V_C, 0, r_N, colors='gray', linestyles='dashed')

r_prime = r_N + L
ax.vlines(r_prime, 0, E_alpha, colors='gray', linewidth=1)

# Labels
ax.text(0.15, 5.2, r'$V(r)$', fontsize=20)
ax.text(-1.0, 1.5, 'Energy', rotation=90, fontsize=18)
ax.text(2, -0.8, 'Radius', fontsize=18)
ax.text(2.8, 2.5, 'Coulomb potential\nenergy', fontsize=16)
ax.text(-0.4, V_C, r'$V_C$', fontsize=9)
ax.text(-0.4, E_alpha, r"$V_C'$", fontsize=9)
ax.text(1.5, E_alpha + 0.15, r'$E_\alpha$', fontsize=18)
ax.text(r_N + 0.02, -0.7, r'$r_N$', fontsize=18)
ax.text(r_prime - 0.5, -1, r"$r' = r_N + L$", fontsize=18)

ax.set_xlim(-0.5, 7)
ax.set_ylim(-6, 6)
ax.set_xticks([])
ax.set_yticks([])

for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```
